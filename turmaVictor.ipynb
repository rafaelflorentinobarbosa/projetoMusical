```python
# Importar o pandas
import pandas as pd
```

```python
df = pd.read_csv('datasets/turmavictor.csv')
```

```python
df.head()
```

```python
pequisa_musical = df.drop(columns=['professor',' faixa_etaria'])

pequisa_musical.head()
```

```python
import matplotlib.pyplot as plt

# Calcular a soma dos votos para cada gênero
soma_votos = pequisa_musical.sum()

# Criar o gráfico de barras horizontais
plt.figure(figsize=(10, 6))
soma_votos.plot(kind='barh', color='skyblue')

# Adicionar título e rótulos
plt.title('Soma dos Votos por Gênero Musical - Turma do Professor Victor')
plt.xlabel('Soma dos Votos')
plt.ylabel('Gênero Musical')

# Exibir o gráfico
plt.tight_layout()
plt.show()
plt.show()

```
